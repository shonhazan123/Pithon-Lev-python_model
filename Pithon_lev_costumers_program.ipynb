```python
import pandas as pd
import glob
import os
import csv
from openpyxl import Workbook
```

```python
files = os.path.join(r'C:\Users\shonh\Desktop\VisualCode\PythonProjects','sales*.csv')
files= glob.glob(files)
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
df = pd.DataFrame(data = df)
df.drop(labels=["סוג פתקית","מספר פתקית","שעת מכירה"],axis=1,inplace=True)
costumers = df.groupby(["לקוח","טלפון לקוח"])["סך מכירה"].sum().reset_index()
df['תאריך מכירה'] = pd.to_datetime(df["תאריך מכירה"])
```

```python
def yearly_avg(df):
    yearly_avg_list=[]
    monthly_avg_list =[]
    new_df = df
    df = df.groupby(["לקוח","טלפון לקוח"])["סך מכירה"].sum().reset_index().sort_values(["סך מכירה"],ascending=False)
    df_list = (new_df.groupby(["לקוח","טלפון לקוח"])["תאריך מכירה"].count().reset_index())["תאריך מכירה"].to_list()
    for costumer in df_list:
        yearly_avg = costumer /30
        monthly_avg = yearly_avg/4
        yearly_avg_list.append(yearly_avg)
        monthly_avg_list.append(monthly_avg)
    df['ממוצע שנתי'] = sorted(yearly_avg_list,reverse = True)
    df['ממוצע חודשי'] = sorted(monthly_avg_list,reverse = True)
    return df
```

```python
df = yearly_avg(df)
```
